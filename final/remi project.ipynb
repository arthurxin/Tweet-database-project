```python
import remi.gui as gui
from remi import start, App
import os
import re
import socket
            
class MyApp(App):
        
    def main(self, name='world'):
        
        host = socket.gethostname()
        self.start_searching_sever(sever_port=(host, 18000))
        
        container = gui.VBox(width=1200, height=600,margin='auto auto')
        self.lbl = gui.Label('')
        container.append(self.lbl)
        
        self.btInputDiag = gui.Button('keyword searching', width=400, height=20, margin='5px')
        self.btInputDiag.onclick.do(self.open_input_dialog)
        container.append(self.btInputDiag)
        
        self.lbl2 = gui.Label('tweet project')
        container.append(self.lbl2)
        
        return container
        
    def open_input_dialog(self, widget):
        self.inputDialog = gui.InputDialog('keyword_searching', '',initial_value='',width=500)
        self.inputDialog.confirm_value.do(self.on_input_dialog_confirm)
        self.inputDialog.show(self)
  
    def on_input_dialog_confirm(self, widget, value):
        server_output = self.sever_connection(value)
        self.lbl.set_text(server_output)
#         self.lbl.set_text("hello"+value)
    
    def sever_connection(self, client_input):
#         inp = input("sending requestï¼š").strip()
        if not client_input:
            return "None return"
        if client_input == "exit":
            return "Bye"
        
        self.sever.sendall(client_input.encode())
        server_output = self.sever.recv(1024).decode()
        
        return server_output
    
    def start_searching_sever(self, sever_port):
        self.sever = socket.socket()
        self.sever.connect(sever_port)
```

```python
start(MyApp, address='0.0.0.0', port=18001)
```

```python
def shutdown_sever():
    cmd = 'netstat -tunlp | grep ' + str(18001)
    f = os.popen(cmd)
    for line in f.readlines():
        line = line.replace("\n", "")
        line = re.sub(' +', ' ', line)
        if line != '':
            res = line.split(' ')
            porcess = res[-2].split("/")[0]
            cmd = 'kill -9 ' + porcess
            os.system(cmd)
```

```python
shutdown_sever()
```
